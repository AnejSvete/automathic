```python
# Example in a Jupyter notebook
import matplotlib.pyplot as plt
import numpy as np
from automathic.algebra.algebra import (
    generate_cyclic_group,
    generate_algebra_mod_n,
    generate_symmetric_group,
)

# Make sure we have the required imports
import networkx as nx
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, HTML

# Create some interesting algebraic structures
Z6 = generate_cyclic_group(6, name="Z6", description="Cyclic group of order 6")
S3 = generate_symmetric_group(3, name="S3", description="Symmetric group on 3 elements")
R4 = generate_algebra_mod_n(4, name="R4", description="Ring of integers mod 4")

# Show basic information about each structure
print(f"Z6 order: {Z6.order}")
print(f"S3 order: {S3.order}")
print(f"R4 order: {R4.order}")

# Simple visualizations
Z6.visualize_cayley_table()
plt.show()

# Visualize the element graph of Z6
Z6.visualize_element_graph()
plt.show()

# Ring-specific visualization
R4.visualize_zero_divisors()
plt.show()
```
