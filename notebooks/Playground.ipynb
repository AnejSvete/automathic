```python
φ.to_fsa(alphabet=["a"]).reindex()
```

```python
from itertools import product

C = SOMtoFSA(alphabet=["a", "b"])

V1 = sorted(φ.get_free_variables())
V2 = sorted(φ.get_free_set_variables())

# Construct the enriched alphabet: pairs of (symbol, variable_set)
# Each pair represents a position in the word with its symbol and variables
V_alphabet = list(
    product(
        C.alphabet,
        C._powerset(V1),
        C._powerset(V2),
    )
)
```

```python
C._construct_V1_structure_fsa(V1, V2, V_alphabet)
```

```python
C._construct_V_structure_fsa(V1, V2, V_alphabet)
```
