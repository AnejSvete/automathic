```python
from example_fsas import (
    create_substring_101_automaton,
    create_end_with_01_automaton,
    create_divisible_by_3_automaton,
)

# Create and display the automaton
fsa = create_divisible_by_3_automaton()

# Test some strings - convert string digits to integers
test_strings = ["", "0", "1", "10", "101", "0101", "1100", "001010"]
for s in test_strings:
    # Convert string digits to integers
    int_input = [int(c) for c in s]
    result = "Accept" if fsa.accepts(int_input) else "Reject"
    print(f"'{s}': {result}")
```

```python
from fo_to_fsa import convert_fo_to_fsa
from parser import parse_fo_formula

# Test the 101 substring formula
formula_101 = "exists x (exists y (exists z (P_1(x) and P_0(y) and P_1(z) and y = x+1 and z = y+1)))"
fsa_101 = convert_fo_to_fsa(formula_101)

# Visualize the generated FSA
print("FSA for strings containing 101:")
fsa_101.trim().minimize().trim()
```
